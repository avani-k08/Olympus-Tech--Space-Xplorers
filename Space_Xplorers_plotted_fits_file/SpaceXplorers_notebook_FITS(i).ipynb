```python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np

fits_file = r"C:\Users\avani\Space_Xplorers\Space_Xplorers_plotted_fits_file\q9.fits"   # change if needed
hdul = fits.open(fits_file)
data = hdul[0].data
hdul.close()

data = np.nan_to_num(data, nan=0.0, posinf=0.0, neginf=0.0)

h, w = data.shape

target_size = 900
y_idx = np.linspace(0, h - 1, target_size)
x_idx = np.linspace(0, w - 1, target_size)

data_resized = data[y_idx.astype(int)[:, None], x_idx.astype(int)[None, :]]

vmin, vmax = np.percentile(data_resized, (1, 99))

fig, ax = plt.subplots(figsize=(9, 9), dpi=100)
img = ax.imshow(data_resized, cmap='plasma', origin='lower', vmin=vmin, vmax=vmax)

cbar = plt.colorbar(img, ax=ax, shrink=0.8)
cbar.set_label('Pixel Intensity', fontsize=12)

ax.set_title('FITS Image', fontsize=14, fontweight='bold')
ax.set_xlabel('X Pixel', fontsize=12)
ax.set_ylabel('Y Pixel', fontsize=12)

for spine in ax.spines.values():
    spine.set_edgecolor('white')
    spine.set_linewidth(2)

plt.tight_layout()
plt.savefig("q9.jpeg", bbox_inches='tight', dpi=100)
plt.show()

print("Saved")
```
